#6. Проверка названий столбцов и замена символов

```python
# Вывод названий столбцов
print("Текущие названия столбцов:")
print(df.columns)

# Замена символов на всякий пожарный
df.columns = df.columns.str.replace('с', 'c')
df.columns = df.columns.str.replace('о', 'o')
df.columns = df.columns.str.replace('е', 'e')
df.columns = df.columns.str.replace('а', 'a')
df.columns = df.columns.str.replace('у', 'y')

# Вывод названий столбцов после изменений
print("\nНовые названия столбцов:")
print(df.columns)
```

Из результата работы метода 'columns' видно, что названия не содержат пробелов и приведены все к нижнему регистру.

Для подстраховки все русские буквы 'а', 'е', 'c', 'у' и 'o' были заменены на английские.

#7. Проверка и устранение пропущенных значений, приведение типов

```python
# Подсчёт пропущенных значений по каждому столбцу
print("Количество пропусков в каждом столбце:")
print(df.isna().sum())

# 1. Строки с пустым названием фильма, годом выпуска
# или длительностью фильма удаляем
df = df.dropna(subset=['title', 'release_year', 'duration'])

# 2. Пустые страны заменим на Unknown
df['main_production'] = df['main_production'].fillna('Unknown')
df['main_genre'] = df['main_genre'].fillna('Unknown')

# 3. В числовых полях заменим неподходящие типы
df['score'] = pd.to_numeric(df['score'], errors='coerce')
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
df['number_of_votes'] = pd.to_numeric(df['number_of_votes'], errors='coerce')
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

# 4. Пропуски в числовых полях заполняем 0
df['score'] = df['score'].fillna(0)
df['number_of_votes'] = df['number_of_votes'].fillna(0)

# Проверка после обработки
print("\nПропуски после обработки:")
print(df.isna().sum())
```

Проверка на пропуски выявила некорректные значения. Все найденные несоответствия и ошибки были успешно обработаны путём удаления некорректных строк при помощи метода 'dropna()' и замены пустых значений методом 'fillna()' с предварительной заменой некоторых типов при помощи метода 'to_numeric()'.

Строки с пустым названием фильма, годом выпуска или длительностью фильма удалены, пустые страны заменены на Unknown, а в числовых полях заменены неподходящие типы и пустые значения заменены на 0.

#8. Проверка на явные и неявные дубликаты

```python
# Проверка на явные дубликаты
print(f"Количество явных дубликатов: {df.duplicated().sum()}")

# Удаление явных дубликатов
df = df.drop_duplicates().reset_index(drop=True)
print(f"Количество дубликатов после удаления: {df.duplicated().sum()}")

# Проверка на неявные дубликаты в столбцах
print("\nПроверка неявных дубликатов в столбце 'main_genre':")
print(df['main_genre'].value_counts())

# Приведение жанров к нижнему регистру
df['main_genre'] = df['main_genre'].str.lower()

# Проверка уникальных значений после приведения
print("\nУникальные жанры после нормализации:")
print(df['main_genre'].unique())

print("\nПроверка неявных дубликатов в столбце 'main_production':")
print(df['main_production'].value_counts())

# Приведение стран к верхнему регистру
df['main_production'] = df['main_production'].str.upper()

print("\nУникальные страны после нормализации:")
print(df['main_production'].unique())
```

Проверка выявила 2 явных дубликата, дубли были удалены при помощи метода 'drop_duplicates()' и проиндексированны при помощи метода 'reset_index(drop=True)'.

Также были выявлены неявные дубликаты в жанре. Дубликаты были устранены путём приведения данных в колонке к нижнему регистру.
